```python
import pandas as pd
import nltk
from nltk.corpus import brown
from nltk import FreqDist
from nltk.corpus import stopwords
```

```python
kw = pd.read_csv("./Keyword Stats 2020-05-13 at 22_32_03 - Keyword Stats 2020-05-13 at 22_32_03.csv")
```

```python
kw.head(10)
```

```python
kw.columns
```

```python
kw.drop(kw.columns.difference(['Keyword','Max search volume','Competition (indexed value)','Top of page bid (low range)','Top of page bid (high range)']), 1, inplace=True)
```

```python
kw.head(10)
```

```python
kw.shape
```

```python
kw['Keyword'] = kw['Keyword'].apply(lambda tokenize: nltk.word_tokenize(tokenize))
```

```python
kw.head(10)
```

```python
stop = stopwords.words('spanish')
kw['Keyword'] = kw['Keyword'].apply(lambda keyword: [word for word in keyword if word not in stop])
```

```python
kw.head(10)
```

```python
freq = pd.Series(' '.join(" ".join(word) for word in kw['Keyword']).split(" ")).value_counts()[:10]
freq
```

```python
import numpy as np

kw['Cat1']= np.where(kw['Keyword'].str.contains('ikea',regex=True),'IKEA','-')
```

```python
kw.head(200)
```

```python
for column in kw['Keyword']:
    if column.str.contains('ikea'):
        kw['Cat_1'] = 'IKEA'
    else: 
        kw['Cat_1'] = ' '
```

```python
kw['Cat__1'] = ['IKEA' if word.str.join(' ').str.contains('ikea') else '-' for word in kw['Keyword']] 
```
