```python
def categories(): 
    mydict = dict()
    while True:
        keyword = input("Keyword:")
        keyword = keyword.strip()
        if keyword == 'q':
            break
        category = input("Category:")
        category = category.strip()
        mydict[keyword] = category
    # print(mydict)
    return mydict
    



if __name__ == '__main__':
    mydict = categories()
    print(mydict)
```
